# Applying train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(car[['name','company','year','kms_driven','fuel_type']],car['Price'],test_size=0.1)
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
ohe=OneHotEncoder()
ohe.fit(car[['name','company','year','kms_driven','fuel_type']][['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
regressor=LinearRegression()
```

# Making a pipeline

```python
pipe=make_pipeline(column_trans,regressor)
```

# Fitting the model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
y_pred
```

```python
y_pred
```

# Checking R2 Score

```python
r2_score(y_test,y_pred)
```

# The best model is found at a certain random state

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(car[['name','company','year','kms_driven','fuel_type']],car['Price'],test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(car[['name','company','year','kms_driven','fuel_type']],car['Price'],test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
